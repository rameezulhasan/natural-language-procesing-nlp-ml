```python
words = ["eating","eats","eaten","writing","writes","programming","programs","final","finalization"]
```

### PorterStemmer

```python
from nltk.stem import PorterStemmer
```

```python
porter_stem = PorterStemmer()
```

```python
for word in words:
    print(word+"---->"+porter_stem.stem(word))
```

```python
porter_stem.stem("congratulations")
```

```python
porter_stem.stem("sitting")
```

### RegexpStemmer

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stem = RegexpStemmer("ing$|e$|s$|able$", min=4)
```

```python
reg_stem.stem("eating")
```

```python
reg_stem.stem("ingeating")
```

```python
reg_stem.stem("decode")
```

### Snowball Stemmer

```python
from nltk.stem import SnowballStemmer
```

```python
snow_stem= SnowballStemmer("english")
```

```python
for word in words:
    print(word+"--->"+snow_stem.stem(word))
```

```python
porter_stem.stem("fairly"),porter_stem.stem("sportingly")
```

```python
snow_stem.stem("fairly"),snow_stem.stem("sportingly")
```
